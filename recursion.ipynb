```python
#eliminate the duplicate data items form the given collection

inp = "HELLO"
def remove_duplicate(coll,i=0,res = ''):
    if i==len(coll):
        return res
    if coll[i] not in res:
        res+=coll[i]
    return remove_duplicate(coll,i+1,res)

remove_duplicate(inp)

```

```python
#WAP to extract only duplicate items from the program

def duplicate(coll,i=0,res='',uni = ''):
    if i == len(coll):
        return res
    if coll[i] not in uni:
        uni+=coll[i]
    else:
        res+=coll[i]
    return duplicate(coll,i+1,res,uni)
duplicate('hello')
```

```python
#Wap to sum of the even values from the collection

coll = '124527'
def sum_even(coll,i=0,sum=0):
    if i==len(coll):
        return sum
    if int(coll[i])%2==0:
        sum+=int(coll[i])
    return sum_even(coll,i+1,sum)
sum_even(coll)
```

```python
# wap to find out the difference of the all odd values from the collection
coll= [1,8,2,-3,5,17]
def odd_diff(coll,i=0,diff=0):
    if i==len(coll):
        return diff
    if int(coll[i])%2!=0:
        diff-=int(coll[i])
    return odd_diff(coll,i+1,diff)
odd_diff(coll)
```

```python
#wap to findout the factorial of the given number

def factorial(num,i=1,res = 1):
    if i==num+1:
        return res
    res *= i
    return factorial(num,i+1,res)
factorial(5)
```

```python
#wap to findout the factorial of the given number
def factorial(n):
    if(n==0) or (n==1):
        return 1 
    else:
        return n * factorial(n-1)
factorial(5)
```

```python
#wap to findout the factorial of the given number
def fact(n,f=1):
    if n==0:
        return f
    f*=n
    return fact(n-1,f)
fact(5)
```

```python
#wap to check whether the given number is prime or not.

def prime(n=2, num=15, fact_count=0):
    if n>num:
        if fact_count==1:
            print("Prime")
        else:
            print("Not prime")
        return
    else:
        if num%n==0:
            fact_count+=1
    prime(n+1,num,fact_count)
prime()
        
    
```

```python
#segregation using recursion

def segragate(st,alpha,numspe,i):
    if i==len(st):
        return alpha,numspe
    if 'a'<=st[i]<='z' or 'A'<=st[i]<='Z':
        alpha+=st[i]
    else:
        numspe+=st[i]
    return segragate(st,alpha,numspe,i+1)
res = segragate("Pyspi123@78$%","","",0)
print(res[0])
print(res[1])
```

```python
#i/p [1,2,3,4,5,6]
#o/p ([1,3,6,10,15,21],56)
#recursion

ol = []
il = [1,2,3,4,5,6]
def sum_list(inputlist,outputlist,sum=0,i=0,t=0):
    if i==len(il):
        return outputlist,t
    sum+=inputlist[i]
    outputlist+=[sum]
    t+=sum
    return sum_list(inputlist,outputlist,sum,i+1,t)
sum_list(il,ol)
```
