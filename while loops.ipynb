```python
#91 WAP to Remove Last Occurred Character in a String.

st = 'HELLOWORLD'
def last_occur(coll,ele):
    res = ''
    i = -1
    count = 0
    while i>(-len(coll)-1):
        if count == 0:
            if ele == coll[i]:
                res+=''
                count+=1
            else:
                res+=coll[i]
        else:
            res+=coll[i]
        i-=1
    return res[::-1]

last_occur(st,'L')
```

```python
#98. WAP to replace a specific character into a new character.

def char_replace(coll,char1, char2):
    i = 0
    res = ''
    while i<len(coll):
        if coll[i]==char1:
            res+=char2
        else:
            res+=coll[i]
        i+=1
    return res

char_replace('hello','l','i')
```

```python
#99. WAP to convert lowercase to uppercase in a given string.

def upper(coll):
    res = ''
    i = 0
    while i<len(coll):
        if 'a' <= coll[i] <= 'z':
            res+=chr(ord(coll[i])-32)
        else:
            res+=coll[i]
        i+=1
    return res

print(upper('Hello'))
```

```python
#100 WAP to convert uppercase to lowercase in a given string.

def lower(coll):
    i = 0
    res = ''
    while i<len(coll):
        if 'A' <= coll[i] <= 'Z':
            res+=chr(ord(coll[i])+32)
        else:
            res+=coll[i]
        i+=1
    return res

lower('HEllo')
```

```python
#101. WAP to convert uppercase to lowercase in even position characters in a given string
def even_lower(coll):
    res = ''
    i = 0
    while i<len(coll):
        if i%2!=0 and 'A' <= coll[i] <= 'Z':
            res+=chr(ord(coll[i])+32)
        else:
            res+=coll[i]
        i+=1
    return res

even_lower('HELLO')
```

```python
#103. WAP to convert the uppercase letter to lowercase letter and lowercase letter to
#uppercase letter, if the character is special symbol replace with * for the given
#input string.

def diff_case(coll):
    i = 0
    res = ''
    while i<len(coll):
        if 'A' <= coll[i] <= 'Z':
            res+=chr(ord(coll[i])+32)
        elif 'a' <= coll[i] <= 'z':
            res+=chr(ord(coll[i])-32)
        else:
            res+= '*'
        i+=1
    return res  

diff_case("H@llO")
```
